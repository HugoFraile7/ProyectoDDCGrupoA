## SCRAPPING

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time


file_path = '/content/Apple_products_part1.csv'
df = pd.read_csv(file_path)
links = df['product_link']

# Función para extraer las reseñas, timestamp y estrellas de un producto.
def get_reviews_from_product(link):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
        "Accept-Language": "en-US, en;q=0.5",
    }
    try:

        response = requests.get(link, headers=headers).        #Realizamos la petición a la página de Amazon.
        soup = BeautifulSoup(response.content, 'html.parser')

        reviews = []
        for review in soup.find_all('div', {'data-hook': 'review'})[:10]:   #Buscaremos la sección de reseñas, y limitamos a 10 reseñas por dispositivo
            review_text = review.find('span', {'data-hook': 'review-body'}).get_text().strip()
            timestamp = review.find('span', {'data-hook': 'review-date'}).get_text().strip()

            # Extraer la calificación en estrellas.
            star_rating_element = review.find('i', {'data-hook': 'review-star-rating'})
            if not star_rating_element:

                # Algunos reviews pueden tener una estructura ligeramente diferente.
                star_rating_element = review.find('i', {'data-hook': 'cmps-review-star-rating'})

            star_rating = star_rating_element.get_text().strip().split(' ')[0]  # Obtiene el valor de las estrellas.

            reviews.append({
                'review': review_text,
                'timestamp': timestamp,
                'stars': star_rating
            })

        return reviews
    except Exception as e:
        print(f"Error al procesar {link}: {e}")
        return []

# Aquí vamos a recopilar nuestros nuevos datos, las reseñas, timestamps y estrellas para cada producto.
all_reviews = []
def scrap(links, nombre):
  for link in links:
      print(f"Scraping reseñas de: {link}")

      # Obtener las 10 reseñas del producto.
      product_reviews = get_reviews_from_product(link)

      # Almacenamos.
      for review in product_reviews:
          all_reviews.append({
              'product_link': link,
              'review': review['review'],
              'timestamp': review['timestamp'],
              'stars': review['stars']
          })


      time.sleep(2)  #


  reviews_df = pd.DataFrame(all_reviews)

  reviews_df.to_csv(nombre, index=False)

  print("Scraping completado.")
```
