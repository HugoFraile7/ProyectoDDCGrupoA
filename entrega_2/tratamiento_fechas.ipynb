```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dataraw/Apple_phones_preprocessed.csv')
df
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['timestamp_nanosec'] = df['timestamp'].astype('int64')
df.head()
```

```python
df['normalized_timestamp_nanosec'] = df.groupby('model_name')['timestamp_nanosec'].transform(lambda x: x - np.min(x))
df.head()
```

```python

# Ordenar los datos por la columna 'timestamp' en orden ascendente
df.sort_values(by='normalized_timestamp_nanosec', ascending=True, inplace=True)

df.head()
```

```python
# Actualizar la columna 'timestamp' con el formato datetime a partir de 'timestamp_nanosec'
df['timestamp_formatted'] = pd.to_datetime(df['normalized_timestamp_nanosec'], unit='ns').dt.strftime('%d/%m/%Y')

df.head()
```
