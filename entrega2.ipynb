```python
import pandas as pd
df = pd.read_csv('dataraw/Apple_phones_merge.csv')
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
# Extract the country and the date from the 'timestamp' column
df['country'] = df['timestamp'].str.extract(r'Reviewed in (.+?) on')[0]
df['timestamp'] = pd.to_datetime(df['timestamp'].str.extract(r'on (.+)')[0], errors='coerce')

# Display the cleaned data to check if the extraction and conversion worked
print(df[['country', 'timestamp']].head())
```

```python
df
```

```python
df.drop(columns = ['ID', 'product_name', 'product_link', 'image_link','brand', 'number_of_ratings', 'rating_out_of_5', 'discount_percentage', 'operating_system', 'screen_size (Inches)','cellular_technology' ], inplace=True)
df
```

```python
# Vemos donde estan situados los datos faltantes
import seaborn as sns
cols = df.columns[:30] # first 30 columns
sns.heatmap(df[cols].isnull())
```

```python
# Vemos el porcentaje de datos faltantes

null_cols = df.isnull().sum()
round(null_cols / df.shape[0] * 100, 2).sort_values(ascending=False)
```

```python
df.drop(columns = ['cpu_model', 'CPU', 'price_before_discount','RAM (GB)', 'available_colors'], inplace=True)
df
```

```python
valores_unicos = df['model_name'].unique()

print(valores_unicos)
```

```python

valores_a_eliminar = ['iPhone', 'IP6S16SLV'] # Cambia estos por los valores que me indiques

df_filtrado = df[~df['model_name'].isin(valores_a_eliminar)]
df = df_filtrado

```

```python
df['model_name'] = df.model_name.str.replace(r'[^\s]13 Pro Max','iPhone 13 Pro Max' )
df['model_name'] = df.model_name.str.replace('Apple iPhone 15 Pro Max','iPhone 15 Pro Max' )
df['model_name'] = df.model_name.str.replace('iPhone 15 Plus (Renewed)','Apple iPhone 15 Plus' )
df['model_name'] = df.model_name.str.replace('Apple iPhone 15 Plus','iPhone 15 Plus' )
df['model_name'] = df.model_name.str.replace('iPhone SE (3rd Gen)','iPhone SE' )

```
