```python
import pandas as pd

# Load the data

data = df

# Extract the date from the 'timestamp' column and convert it to a datetime format
# The original format is "Reviewed in the United States on <date>", so we extract the date part
data['timestamp'] = pd.to_datetime(data['timestamp'].str.extract(r'on (.+)')[0], errors='coerce')

# Display the cleaned data to check if the conversion worked
print(data[['timestamp']].head())

```

```python
# Plotting
import pandas as pd
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))

# Scatter plot for each model name with a different color
for model in data['model_name'].unique():
    model_data = data[data['model_name'] == "iPhone 13 Pro Max"]
    plt.scatter(model_data['timestamp'], model_data['stars'], label=model, alpha=0.7)

# Formatting the plot
plt.title('Rating Over Time by Model')
plt.xlabel('Timestamp')
plt.ylabel('Rating (out of 5)')
plt.legend(title='Model Name')
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
